```python
# 載入需要的套件
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns # 另一個繪圖-樣式套件
plt.style.use('ggplot')

# 忽略警告訊息
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
"""
Your Code Here
"""
matrix = 2*(np.random.random_sample((10, 10))-0.5) # [0.0,1)>[-0.5,0.5)>[-1,1)
plt.figure(figsize=(10,10))
"""
Your Code Here
"""
sns.heatmap(matrix, cmap = plt.cm.RdYlBu_r, vmin = -0.25, annot = True , vmax = 0.6)
plt.show()
```

```python
nrow = 1000
ncol = 3
"""
Your Code Here
"""
matrix = 2*(np.random.random_sample((1000, 3))-0.5)

# 隨機給予 0, 1, 2 三種標籤
indice = np.random.choice([0,1,2], size=nrow) #1D array
plot = pd.DataFrame(matrix, indice)
plot_data = pd.DataFrame(matrix, indice).reset_index() # 將index 變成columes
print(plot_data)
print(plot)

# 繪製 seborn 進階 Heatmap
grid = sns.PairGrid(data = plot_data, size = 3, diag_sharey=False,  
                    hue = 'index', vars = [x for x in list(plot_data.columns) if x != 'index']) #hue : index name ; vars : columes name
#diag_sharey 設成 True 的話，表示對角線上都會參考相同的 Y 座標值。


grid.map_upper(plt.scatter , alpha = 0.2)
grid.map_diag(sns.kdeplot)
grid.map_lower(sns.kdeplot , cmap = plt.cm.OrRd_r)

plt.show()
```
